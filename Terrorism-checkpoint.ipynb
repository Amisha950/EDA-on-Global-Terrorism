```python
plt.subplots(figsize=(15, 8))
sns.countplot(x='iyear', data=df, edgecolor=sns.color_palette('dark', 7))
plt.xticks(rotation=90)
plt.xlabel('Year of attack')
plt.title('Number Of Terrorist Activities Each Year')
plt.show()
```

```python
#incident count per month(overall)
plt.figure(figsize=(15,7))
sns.countplot(data=df,x='imonth')
```

```python
df['attacktype1_txt'].value_counts()
```

```python
plt.subplots(figsize=(15,4))
sns.countplot(x='attacktype1_txt',data=df,palette='inferno',order=df['attacktype1_txt'].value_counts().index)
plt.xticks(rotation=90)
plt.xlabel('Attack Type')
plt.title('Attacking Methods by Terrorists')
plt.show()
```

```python
plt.subplots(figsize=(15,5))
sns.countplot(x='country_txt',data=df,palette='inferno',order=df['country_txt'].value_counts()[:15].index)
plt.xticks(rotation=90)
plt.xlabel('Countries')
plt.title('Number Of Terrorist Activities By Countries')
plt.show()
```

```python
# Calculate the number of cases by country
cases_by_country_df = df['country_txt'].value_counts().reset_index()
cases_by_country_df.columns = ['country_txt', 'Number of cases']

# Create a choropleth plot
px.choropleth(
    cases_by_country_df,
    locations='country_txt',
    color='Number of cases',
    locationmode='country names',
    hover_name='country_txt',
    projection='natural earth'
)
```

```python
#count of cases in cities of Iraq
iraq_df=df[df['country_txt']=='Iraq']
iraq_df['city'].value_counts()
```

```python
plt.subplots(figsize=(15,5))
sns.countplot(x='city',data=iraq_df,palette='inferno',order=iraq_df['city'].value_counts()[:15].index)
plt.xticks(rotation=90)
plt.xlabel('Cities in Iraq')
plt.title('Number Of Terrorist Activities By Cities in Iraq')
plt.show()
```

```python
#plot of target types by terrorists in Iraq
plt.subplots(figsize=(15,5))
sns.countplot(x='targtype1_txt',data=iraq_df,palette='inferno',order=iraq_df['targtype1_txt'].value_counts()[:15].index)
plt.xticks(rotation=90)
plt.xlabel('Target type in Iraq')
plt.title('Number Of Target types in Iraq')
plt.show()
```

```python
#plot of weapon types used by terrorists in Iraq
plt.subplots(figsize=(13,5))
sns.countplot(x='weaptype1_txt',data=iraq_df,palette='inferno',order=iraq_df['weaptype1_txt'].value_counts().index)
plt.xticks(rotation=90)
plt.xlabel('Weapon type used in Iraq')
plt.title('Number Of Weapon types used in Iraq')
plt.show()
```
